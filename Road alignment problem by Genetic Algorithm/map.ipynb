```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt

#**********
path = './map.npy'

# Genetic Algorithm Parameters
population_size = 300
mutation_rate = 0.7
crossover_rate = 1.0
max_generations = 1000

# Road Map Parameters
map_size = (30,75)
roughness_levels = [1, 2, 3, 4, 5, 6, 7]
roughness_probabilities = [0.7, 0.15, 0.09, 0.02, 0.02, 0.01, 0.01]

# Create a random road map
def create_road_map():
    road_map = np.random.choice(roughness_levels, size=map_size, p=roughness_probabilities)
    return road_map

# Calculate the cost of building a road between two coordinates
def calculate_cost(coord1, coord2):
    if abs(coord1[0] - coord2[0]) <= 1 and abs(coord1[1] - coord2[1]) <= 1:
        return road_map[coord2[0], coord2[1]]
    else:
        distance = (coord1[0] - coord2[0])**2 + (coord1[1] - coord2[1])**2 + road_map[coord2[0], coord2[1]]
        return distance

# Calculate the fitness value of an individual
def calculate_fitness(individual):
    total_cost = 0
    for i in range(len(individual) - 1):
        total_cost += calculate_cost(individual[i], individual[i+1])
    return total_cost

# Generate initial population
def generate_population():
    population = []
    for _ in range(population_size):
        individual = [(random.randint(0, map_size[0]-1), i) for i in range(map_size[1])]
        #random.shuffle(individual)
        population.append(individual)
    return population

# Perform bitwise mutation on an individual
def mutate(individual):
    for i in range(len(individual)):
        if random.random() < mutation_rate:
            individual[i] = (random.randint(0, map_size[0]-1), individual[i][1])
    return individual

#******** random mutation

def mutate1(individual):
    if random.random() < mutation_rate:
        i = random.randint(0, map_size[1]-1)
        individual[i] = (random.randint(0, map_size[0]-1), individual[i][1])
    return individual

# Perform crossover between two individuals
def crossover(parent1, parent2):
    child = parent1.copy()
    for i in range(len(parent2)):
        if random.random() < crossover_rate:
            child[i] = parent2[i]
    return child

#********* 1-point crossover

def crossover1(parent1, parent2):
    child = parent1.copy()
    if random.random() < crossover_rate:
        i = random.randint(1, map_size[1]-2)
        child[i:] = parent2[i:]
    return child

# Select parents for reproduction
def select_parents(population):
    parents = []
    for _ in range(2):
        tournament = random.choices(population, k=3)
        tournament_fitness = [calculate_fitness(individual) for individual in tournament]
        winner_index = tournament_fitness.index(max(tournament_fitness))
        parents.append(tournament[winner_index])
    return parents

# Perform survival selection to generate the next generation
def select_survivors(population):
    sorted_population = sorted(population, key=lambda x: calculate_fitness(x))
    survivors = sorted_population[:population_size]
    return survivors

#******* tournament survivor

def select_survivors1(population):
    survivors = []
    while len(survivors) < population_size:
        tournament = random.sample(population, 3)
        winner = min(tournament, key=lambda x: calculate_fitness(x))
        survivors.append(winner)
    return survivors

# Run the genetic algorithm
def run_genetic_algorithm():
    population = generate_population()
    best_fitness_values = []
    average_fitness_values = []

    for generation in range(max_generations):
        parents = select_parents(population)
        offspring = [crossover(parents[0], parents[1]) for _ in range(population_size)]
        offspring = [mutate1(individual) for individual in offspring]
        population = select_survivors(population + offspring)

        best_fitness = min([calculate_fitness(individual) for individual in population])
        average_fitness = sum([calculate_fitness(individual) for individual in population]) / population_size

        best_fitness_values.append(best_fitness)
        average_fitness_values.append(average_fitness)

        print(f"Generation {generation+1}: Best Fitness = {best_fitness}, Average Fitness = {average_fitness}")

    best_individual = min(population, key=calculate_fitness)
    x_coords = [coord[1] for coord in best_individual]
    y_coords = [coord[0] for coord in best_individual]


    fig, ax = plt.subplots()
    ax.imshow(road_map, cmap='gray', origin='lower')
    ax.plot(x_coords, y_coords, linewidth=2, color='red')
    ax.set_title('Optimal Road Path')
    ax.set_xlabel('x-coordinate')
    ax.set_ylabel('y-coordinate')
    plt.show()

    plt.plot(best_fitness_values, label='Best Fitness')
    plt.plot(average_fitness_values, label='Average Fitness')
    plt.xlabel('Generation')
    plt.ylabel('Fitness')
    plt.legend()
    plt.show()

# Main driver code
if __name__ == '__main__':
    road_map = np.load(path)
    #create_road_map()
    map_size = tuple(road_map.shape)
    run_genetic_algorithm()
```
